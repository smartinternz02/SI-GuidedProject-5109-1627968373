```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([("one",OneHotEncoder(),[3])],remainder = "passthrough")
i = ct.fit_transform(i)
```

```python
from sklearn.model_selection import train_test_split
i_train,i_test,o_train,o_test = train_test_split(i,o,test_size = 0.2,random_state = 0)
```

```python
print(i.shape,"input shape ")
print(o.shape,"output shape ")
print(i_train.shape,"input_train shape ")
print(i_test.shape,"input_test shape ")
print(o_train.shape,"output_train shape ")
print(o_test.shape,"output_test shape ")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
i_train = s.fit_transform(i_train)
i_test = s.fit_transform(i_test)
```

# ANN modeling on Startup dataset

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
startup = Sequential()
```

```python
startup.add(Dense(units= 6 , kernel_initializer ="random_uniform", activation = "relu"))
```

```python
startup.add(Dense(units= 18 , kernel_initializer ="random_uniform", activation = "relu")) #hidden layer1
```

```python
startup.add(Dense(units= 18 , kernel_initializer ="random_uniform", activation = "relu")) #hidden layer2
```

```python
startup.add(Dense(units= 18 , kernel_initializer ="random_uniform", activation = "relu")) #hidden layer3
```

```python
startup.add(Dense(units= 1 , kernel_initializer ="random_uniform", activation = "relu"))
```

```python
startup.compile(optimizer="rmsprop", loss="mse" ,metrics=["mse"])
```

```python
startup.fit(i_train,o_train, epochs = 60000)
```

```python
ypred = startup.predict(i_test)
```
